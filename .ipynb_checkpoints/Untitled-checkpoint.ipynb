```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
#firstly I created dataFrame structure
#data info
df = pd.DataFrame({
    'x': iris.data[:,0],
    'y': iris.data[:,1],
    'cluster' : iris.target
})
df.info()
df
```

```python
centroids = {}
for i in range(3):
    llist = []
    llist.append(df.loc[df['cluster'] == i]['x'].mean())
    llist.append(df.loc[df['cluster'] == i]['y'].mean())
    
    centroids[i] = llist
centroids
```

```python
#Question1
#showing where each data instance falls on the axes relating to Sepal Width and Sepal Length plot
fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], color='darkviolet')
plt.title("Length vs Width")
plt.show
```

```python
#Question2
#plotting points
#color code the plot
fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], c=iris.target)
plt.xlabel('Speal Length', fontsize=14)
plt.ylabel('Sepal Width', fontsize=14)
plt.grid()
```

```python
#Question3 
#defining the colors and ploting centroids for all clusters
#It also works with clusters 1,3,5
fig = plt.figure(figsize=(4, 3))
#I defined the figure size
colmap = {0: 'r', 1: 'g', 2: 'b'} 
#defined the clusters color code
#plot the initial random  cluster centers for k=3
for i in range(3):
    plt.scatter(centroids[i][0],centroids[i][1], color=colmap[i])
plt.show()
```

```python
#3rd question continuation 
#showing the complete graph
fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], c=iris.target,alpha = 0.3)
colmap = {0: 'r', 1: 'g', 2: 'b'}
col = [0,1]
for i in centroids.keys():
    plt.scatter(centroids[i][0],centroids[i][1], c=colmap[i], edgecolor='k')
plt.show()
```

```python
#placement function: Calculating distance and updating dataFrame
#we assign each point to the closest cluster centroid
def placement(df, centroids):
    for i in range(3):
        #Euclidian distance to find the distance between any two data points
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df['x'] - centroids[i][0]) ** 2
                + (df['y'] - centroids[i][1]) ** 2
            )
        )
    centroiddistancecols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroiddistancecols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df
```

```python
df = placement(df, centroids)
df
```

```python
#Visualisation of clusters with colors:
fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.3)
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i], edgecolor='k')

plt.show()
```

```python
#Question4
# the new cluster centers for each iteration of the k-means algorithm until it converges
#Update function: Updating the centroids
def update(k):
    for i in range(3):
        centroids[i][0] = np.mean(df[df['closest'] == i]['x'])
        centroids[i][1] = np.mean(df[df['closest'] == i]['y'])
    return k
```

```python
centroids = update(centroids)
centroids
```

```python
#4th question continuation 
#Visualising the graph with updated centroids
fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.3)
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i], edgecolor='k')

plt.show()
```

```python
#Repeat assignment stage to to ressign the points with clusters
df = placement(df, centroids)
```

```python
#Visualising the updated graph
fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], color=df['color'], alpha=0.3)
for i in centroids.keys():
    plt.scatter(centroids[i][0],centroids[i][1], color=colmap[i], edgecolor='k')
plt.show()
```

```python
#4th question continuation 
#Continuing until all assigned clusters don't change any more
while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = placement(df, centroids)
    if closest_centroids.equals(df['closest']):
        break
```

```python
#final result
#Cluster assignments and cluster centers found by k-means with three clusters

fig = plt.figure(figsize=(4, 3))
plt.scatter(df['x'], df['y'], color=df['color'])
for i in centroids.keys():
    plt.scatter(centroids[i][0],centroids[i][1], color=colmap[i], edgecolor='k')

plt.show()
```

```python
#Question5
#In the first plot, it was observed that while the data points were mixed with each other, they were separated with sharper boundaries in the cluster one
#clustering is somewhat similar to classification, in that each item gets a label
#However, there is no ground truth
#We can see that the compact points have been assigned to a single cluster. Whereas the points that are spread loosely but were in the same cluster, have been assigned to different clusters
plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.scatter(df['x'], df['y'], c=iris.target)
plt.xlabel('Spea1 Length', fontsize=8)
plt.ylabel('Sepal Width', fontsize=8)
plt.show()
plt.figure(figsize=(8,3))
plt.subplot(1,2,2)
plt.scatter(df['x'], df['y'],color=df['color'])
for i in centroids.keys():
    plt.scatter(centroids[i][0],centroids[i][1], color=colmap[i], edgecolor='k')

plt.show()
```
